```python
# Simplified implementation of Queue (using built-ins)

class QueueII:
    
    class __Node:
        
        def __init__(self, data):
            self.data = data
            self.next = None
        

    
    def __init__(self):
        #self.items =[]
        self.front = None
        self.rear = None
        self.length = 0

    def enqueue(self,value):
        #create a new node
        #check the queue's status to see if it is emty: if so set front and rear to the new node
        #else set the next value of the rear of the queue to the new node, and set the rear of the queue to the new node
        new_node = self.__Node(value)
        if self.length == 0:
            self.front = new_node
            self.rear = new_node
            self.length +=1
        else:
            self.rear.next = new_node
            self.rear = new_node
            self.length +=1
        #self.items.insert(0,value)

    def dequeue(self):
        #if the front and rear are equal then empty the queue
        #else set the front equal to the front.next
        if self.length == 0:
            raise IndexError("queue is empty")
        if self.front == self.rear:
            datum = self.front.data
            self.front = None
            self.rear = None
            self.length = 0

            return datum 
            
        else:
            datum = self.front.data
            self.front = self.front.next
            self.length -=1
            return datum
        
       # return self.item.pop()


    #nice to have methods
    def size(self):
        return self.length

    def peek(self):
        return self.front.data

    def is_empty(self):
        return self.front.data

```
